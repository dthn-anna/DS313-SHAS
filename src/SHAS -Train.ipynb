```
import os
import glob
import subprocess
from multiprocessing import Pool

# Bước 1: Lấy đường dẫn gốc
MTEDX_ROOT = os.environ["MTEDX_ROOT"]

# Bước 2: Tìm tất cả các file .flac
flac_files = glob.glob(f"{MTEDX_ROOT}/*/data/*/wav/*.flac")
print(f"Found {len(flac_files)} FLAC files.")

# Bước 3: Hàm xử lý từng file
def convert_and_remove(flac_file):
    wav_file = flac_file.replace(".flac", ".wav")
    try:
        result = subprocess.run([
            "ffmpeg", "-i", flac_file,
            "-ac", "1", "-ar", "16000",
            "-hide_banner", "-loglevel", "error",
            wav_file
        ], check=True)

        # Nếu chuyển đổi thành công → xóa file gốc
        if os.path.exists(wav_file):
            os.remove(flac_file)
            print(f"✅ Converted and removed: {flac_file}")
        else:
            print(f"⚠️  Conversion failed: {flac_file} (no .wav created)")
    except subprocess.CalledProcessError as e:
        print(f"❌ ffmpeg error on {flac_file}: {e}")

# Bước 4: Chạy song song với 12 tiến trình
with Pool(12) as pool:
    pool.map(convert_and_remove, flac_files)
```
